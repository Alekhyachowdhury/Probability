```python

import pandas as pd
import numpy as np
from numpy import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dice = list(np.arange(-1,2,2))

num_iter_list =[5000,500,100,10]
iteration_count = len(num_iter_list)
fig, axes = plt.subplots(iteration_count,figsize=(8, 12))
fig.subplots_adjust(hspace=0.6, wspace=0.2)
sns.set_theme(style="whitegrid", palette="pastel")
sns.set(font_scale=.7)
index = 0
for iteration in num_iter_list:
  #print(f'start interation {iteration}')
  n = 0
 
  sum_list = []
  while n < iteration:
    x1 = random.choice(dice)
    x2 = random.choice(dice)
    data = f'dice1 value {x1} dice2 value {x2}'
    sum_value = {"sum": x1 + x2}
    sum_list.append(sum_value)
    n = n+1
  df = pd.DataFrame(sum_list)

  axes[index].set_title(f'Interation count :- {iteration}')
  ax = sns.countplot(df,x="sum",ax=axes[index],width=0.5)
  ax.bar_label(axes[index].containers[0])

  index = index + 1

plt.show()
```
